```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
HIV = pd.read_csv('IHME-GBD_2021_DATA-efa25464-1.csv')
HIV
```

```python
HIV = HIV[HIV['metric_id'] == 3]
HIV 
```

```python
HIVdeaths = HIV[HIV['measure_name'] == 'Deaths']
HIVYLDs = HIV[HIV['measure_name'] == 'YLDs (Years Lived with Disability)']
HIVPrevalence = HIV[HIV['measure_name'] == 'Prevalence']
HIVIncidence = HIV[HIV['measure_name'] == 'Incidence']
```

```python
HIVdeaths
```

```python
HIVdeaths['year'].nunique()
```

```python
HIVdeaths['location_name'].nunique()
```

```python
HIVdeathsvis = HIVdeaths[['location_name', 'year', 'val']].groupby(by=['year', 'location_name']).sum().values
HIVYLDsvis = HIVYLDs[['location_name', 'year', 'val']].groupby(by=['year', 'location_name']).sum().values
HIVPrevalencevis = HIVPrevalence[['location_name', 'year', 'val']].groupby(by=['year', 'location_name']).sum().values
HIVIncidencevis = HIVIncidence[['location_name', 'year', 'val']].groupby(by=['year', 'location_name']).sum().values
```

```python
HIVdeathsvis
```

```python
HIVdeathsvis.shape
```

```python
locationdeaths = HIVdeaths['location_name'].unique()
locationYLDs = HIVYLDs['location_name'].unique()
locationPrevalence = HIVPrevalence['location_name'].unique()
locationIncidence = HIVIncidence['location_name'].unique()
```

```python
HIVdeathsrange = HIVdeaths['year'].nunique(), HIVdeaths['location_name'].nunique()
HIVYLDsrange = HIVYLDs['year'].nunique(), HIVYLDs['location_name'].nunique()
HIVPrevalencerange = HIVPrevalence['year'].nunique(), HIVPrevalence['location_name'].nunique()
HIVIncidencerange = HIVIncidence['year'].nunique(), HIVIncidence['location_name'].nunique()
HIVYLDsrange
```

```python
HIVdeathsdf = pd.DataFrame(HIVdeathsvis.reshape(HIVdeathsrange), index = list(range(1980, 2022)), columns = locationdeaths)
HIVYLDsdf = pd.DataFrame(HIVYLDsvis.reshape(HIVYLDsrange), index = list(range(1990, 2022)), columns = locationYLDs)
HIVPrevalencedf = pd.DataFrame(HIVPrevalencevis.reshape(HIVPrevalencerange), index = list(range(1990, 2022)), columns = locationPrevalence)
HIVIncidencedf = pd.DataFrame(HIVIncidencevis.reshape(HIVIncidencerange), index = list(range(1990, 2022)), columns = locationIncidence)
HIVYLDsdf
```

```python
fig, ax = plt.subplots(figsize = (30, 12))
HIVYLDsdf.plot(kind='line', ax=ax)
plt.show()
```

```python
HIVdeathsdf.to_excel('visual_HIVdeaths_rate.xlsx')
HIVYLDsdf.to_excel('visual_HIVYLDs_rate.xlsx')
HIVPrevalencedf.to_excel('visual_HIVPrevalence_rate.xlsx')
HIVIncidencedf.to_excel('visual_HIVIncidence_rate.xlsx')
```
